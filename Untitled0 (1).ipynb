Cell 1 — Import libraries

This cell loads the Python packages needed to upload files in Colab, process images (OpenCV and NumPy), perform skeletonization (scikit-image), and display results (matplotlib).

```python
from google.colab import files
import cv2
import numpy as np
from skimage.morphology import skeletonize
from skimage.util import img_as_ubyte
import matplotlib.pyplot as plt
```

Cell 2 — Preprocessing function

This cell defines a function that converts a color fingerprint image to grayscale, enhances contrast with CLAHE, smooths with Gaussian blur, thresholds it to a binary image, inverts the binary mask so ridges are foreground, converts to a boolean image for skeletonization, extracts the skeleton (thin ridge lines), and returns the skeleton as an 8-bit image ready for feature extraction.

```python
def preprocess(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    cl = clahe.apply(gray)
    blur = cv2.GaussianBlur(cl, (5,5), 0)
    _, th = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    inv = cv2.bitwise_not(th)
    bin_img = (inv > 0).astype(np.uint8)
    ske = skeletonize(bin_img)
    ske_u = img_as_ubyte(ske)
    return ske_u
```

Cell 3 — Feature extraction and comparison functions

This cell defines two functions: one to extract ORB keypoints and descriptors from a preprocessed image, and another to compare two preprocessed fingerprint images by matching their descriptors with a brute-force Hamming matcher and returning the average match distance as a similarity score (lower = more similar).

```python
def extract_features(img):
    orb = cv2.ORB_create(nfeatures=1000)
    kp, des = orb.detectAndCompute(img, None)
    return kp, des

def compare_fingerprints(img1, img2):
    kp1, des1 = extract_features(img1)
    kp2, des2 = extract_features(img2)
    if des1 is None or des2 is None:
        return 0
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    matches = bf.match(des1, des2)
    if not matches:
        return 0
    scores = [m.distance for m in matches]
    avg_score = np.mean(scores)
    return avg_score
```

Cell 4 — Upload and preprocess first fingerprint (Person A – Image 1)

This cell prompts you to upload the first fingerprint image for Person A, reads it into OpenCV, and runs the preprocessing function to produce the skeleton image used for matching.

```python
print("Upload first fingerprint (Person A - Image 1)")
uploaded = files.upload()
img1_path = list(uploaded.keys())[0]
img1 = cv2.imread(img1_path)
proc1 = preprocess(img1)
```

Cell 5 — Upload and preprocess second fingerprint (Person A – Image 2)

This cell prompts you to upload a second fingerprint image belonging to the same person (Person A), reads and preprocesses it, then computes and prints the similarity score between the two samples from Person A.

```python
print("Upload second fingerprint (Person A - Image 2)")
uploaded = files.upload()
img2_path = list(uploaded.keys())[0]
img2 = cv2.imread(img2_path)
proc2 = preprocess(img2)

score_same = compare_fingerprints(proc1, proc2)
print("Similarity score (same person):", score_same)
```

Cell 6 — Upload and preprocess third fingerprint (Different person)

This cell prompts you to upload a third fingerprint image from a different person, reads and preprocesses it, then computes and prints the similarity score between the first sample of Person A and this third sample (expected to be higher if not matching).

```python
print("Upload third fingerprint (Different person)")
uploaded = files.upload()
img3_path = list(uploaded.keys())[0]
img3 = cv2.imread(img3_path)
proc3 = preprocess(img3)

score_diff = compare_fingerprints(proc1, proc3)
print("Similarity score (different person):", score_diff)
```

Cell 7 — Display results and verdicts

This cell defines a threshold for deciding matches (lower average distance = stronger match), converts scores to MATCHED or NOT MATCHED labels, and displays the three original images side-by-side with the match results clearly shown in the subplot titles.

```python
threshold = 40
result_same = "MATCHED ✅" if score_same < threshold else "NOT MATCHED ❌"
result_diff = "MATCHED ✅" if score_diff < threshold else "NOT MATCHED ❌"

plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("Person A - Image 1")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title("Person A - Image 2\nResult: " + result_same)
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.title("Different Person\nResult: " + result_diff)
plt.axis("off")

plt.tight_layout()
plt.show()
```
